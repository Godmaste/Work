```python
from ipywidgets import VBox, jsdlink, IntSlider


s1, s2 = IntSlider(max=200, value=100), IntSlider(value=40)
jsdlink((s1, 'value'), (s2, 'max'))
connectedsliders = VBox([s1, s2])
```

```python
l = list(open("z.txt"))

dp1 = widgets.Text(value = "", description = "Search: ")
dp2 = widgets.Dropdown(options = l)
combo = widgets.VBox([dp1, dp2])
def combobox(x):
    global dp1
    global dp2
    dp2.options = l
    search = dp1.value
    dp2.index = None #This line is very important for setting the values for widgets other than widget with observe method
    dp2.index = 0
    output = []
    for j in dp2.options:
        if search in j:
            output.append(j)
    if len(output) == 0:
        dp2.options = ["No Items Found"]
    else:
        dp2.options = output
            
dp1.observe(combobox)
```

```python
panel = widgets.Output()
with panel:
    create_panel([[VBox([connectedsliders, combo]), table], [chart]], #widgets
                   [[0, 0], [1]], #stretch 
                   [['top', 0], [0]], #vertical align
                   ['left', 'middle'], #horizontal align
                   0, 0, #height width (0, 0 for default)
                   [[[0,0],[0,0]], [[0,0]]] #widget [height, width]
                  )
```
