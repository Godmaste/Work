```python
from satin import *
pd.set_option('display.latex.repr', True)
pd.set_option('display.latex.longtable', True)
```

```python
file = open("image.png", "rb")
image = file.read()

chart = widgets.Image(
    value=image,
    format='png',
)
```

```python
html = widgets.HTML(
    value='<h3>Thematic Discovery</h3>'
    '<text>Complaints will be sorted on the basis of theme.</text>'
    '<br>'
    '<text>Please type in the specific seed phrase or keywords to get started.</text>'
    '<br>'
    '<br>'
)
```

```python
table = widgets.Output()
with table:
    display(pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                          "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                      "small", "large", "small", "small",
                         "large"],
                 "D": [1, 2, 2, 3, 3, 4, 5, 6, 7]}))
```

```python
l = list(open("z.txt"))

dp1 = widgets.Text(value = "", description = "Search: ")
dp2 = widgets.Dropdown(options = l)
combo = widgets.VBox([dp1, dp2])
def combobox(x):
    global dp1
    global dp2
    dp2.options = l
    search = dp1.value
    dp2.index = None #This line is very important for setting the values for widgets other than widget with observe method
    dp2.index = 0
    output = []
    for j in dp2.options:
        if search in j:
            output.append(j)
    if len(output) == 0:
        dp2.options = ["No Items Found"]
    else:
        dp2.options = output
            
dp1.observe(combobox)
```

```python
create_panel([[VBox([html, combo]), table], [chart]], #widgets
                   [[0, 0], [1]], #stretch 
                   [['top', 0], [0]], #vertical align
                   ['left', 'middle'], #horizontal align
                   0, 0, #height width (0, 0 for default)
                   [[[0,0],[0,0]], [[0,0]]] #widget [height, width]
                  )
```
